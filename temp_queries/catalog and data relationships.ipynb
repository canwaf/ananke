```python
PREFIX dcat: <http://www.w3.org/ns/dcat#>
PREFIX dcterms: <http://purl.org/dc/terms/>
PREFIX owl: <http://www.w3.org/2002/07/owl#>
PREFIX qb: <http://purl.org/linked-data/cube#>
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX skos: <http://www.w3.org/2004/02/skos/core#>
PREFIX void: <http://rdfs.org/ns/void#>
PREFIX xsd: <http://www.w3.org/2001/XMLSchema#>

SELECT ?dataset ?cntObs (COUNT(?dimension) as ?cntDim)
WHERE {
	{
      SELECT ?dataset (COUNT(?observation) as ?cntObs) 
      WHERE { 

      ?dataset a qb:DataSet.
      ?observation qb:dataSet ?dataset.

      }
      GROUP BY ?dataset
      ORDER BY ASC(?cntObs)
      LIMIT 100
	}

 ?dataset qb:structure/qb:component/qb:dimension ?dimension
} 
GROUP BY ?dataset ?cntObs 
ORDER BY ASC(?cntObs) ASC(?cndDim)
```
